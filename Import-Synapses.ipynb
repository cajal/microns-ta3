## Import synapses

```python
import ta3
import pandas as pd
from tqdm import tqdm
```

```python
d = pd.read_csv('full_final_edges_sem_trimmed.csv')
```

```python
synapse = ta3.Synapse()
for _, row in tqdm(d.iterrows(), total=len(d)):
    synapse.insert1(dict(
        segmentation=1, 
        synapse_id=row['ids'], 
        presyn=row['segs_1'], 
        postsyn=row['segs_2'],
        synapse_x=row['locs_1'],
        synapse_y=row['locs_2'],
        synapse_z=row['locs_3'],
        syn_bbox_x1=row['bboxes_1'],
        syn_bbox_y1=row['bboxes_2'],
        syn_bbox_z1=row['bboxes_3'],
        syn_bbox_x2=row['bboxes_4'],
        syn_bbox_y2=row['bboxes_5'],
        syn_bbox_z2=row['bboxes_6']
    ))
```
